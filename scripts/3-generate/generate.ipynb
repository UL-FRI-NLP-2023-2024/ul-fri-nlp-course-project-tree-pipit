```python
grouped = df.groupby('index')

result = []
for name, group in grouped:
    if len(group) < 2:
        continue

    first = group.iloc[0]
    second = group.iloc[1]

    new_row = {
        'instruction': f"{first['title']} - {first['text']}".strip(),
        'input': '',
        'output': second['text']
    }

    result.append(new_row)
```

```python
instruct_1_df = pd.DataFrame(result)
with open('instruct_1.json', 'w', encoding='utf-8') as f:
    for index, row in instruct_1_df.iterrows():
        json.dump(row.to_dict(), f)
        f.write('\n')
```
